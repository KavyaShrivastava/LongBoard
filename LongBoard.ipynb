```r

vars = c("Avg_Speed", "Distance", "Avg_HR", "Calories", "Experience", "Wheel_Durometer", "Ascent_Ratio", "Total_Ascent", "Date", "Temp", "Wind_Speed", "Board_Type", "Best_Lap_Time", "Number_of_Laps")
preprocess <- function(final_data){
    pro_df <- final_data
    first_day <- "2021-03-13 00:00:00"
    pro_df$Experience <- log(as.numeric(difftime(pro_df$Date, first_day, units = "days")))
    pro_df$Ascent_Ratio <- pro_df$Total_Descent/pro_df$Total_Ascent
    pro_df$Wheel_Ratio <- pro_df$Wheel_Diameter/pro_df$Wheel_Durometer
    pro_df <- subset(pro_df, select = vars)
    return(pro_df)
    }
```

```r
# df
pro_df <- preprocess(final_data)
head(pro_df, 10)
```

```r
get_best_result = function(caret_fit) {
  best = which(rownames(caret_fit$results) == rownames(caret_fit$bestTune))
  best_result = caret_fit$results[best, ]
  rownames(best_result) = NULL
  best_result
}
```

We observe from the lmg that board category explains roughtly 2.8%. Board Type explains 2.5%, wheel diameter explains 2.6%, wheel durometer explains 3.2%.
